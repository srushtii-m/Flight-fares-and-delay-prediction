```python
import pandas as pd
from datetime import datetime
```

```python
df = pd.read_csv('indigodelay_1000_new.csv')
df.head()
df_blah=pd.read_csv('srushtiadd.csv')
df_blah.head()
df = pd.merge(df, df_blah, on=['text'])
```

```python
df
```

```python
#converting into data time object
df['date'] = pd.to_datetime(df['date'], format='%Y-%m-%d')
print(df.info())
```

```python
print(df['date'].min())
```

```python
#df.assign(Count=1).groupby('date')['Count'].count()
```

```python
df['tweet_volume'] = df.groupby('date')['date'].transform('count')
print(df)
```

```python
df1=df
df1.drop(['id', 'created_at','favorite_count','retweet_count','text','entities','screen_name'], axis=1,inplace=True)
```

```python
df1.drop_duplicates(inplace=True)
```

```python
df1.to_csv('tweetvolume.csv', encoding='utf-8', index=False)
```
